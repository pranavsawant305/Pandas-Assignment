### Step 4. How many products cost more than $10.00? 

Use `str` attribute to remove the $ sign and convert the column to proper numeric type data before filtering.


```python
df=df["order_id\tquantity\titem_name\tchoice_description\titem_price"].str.split("\t",expand=True)
df.columns=["order_id","quantity","item_name","choice_description","item_price"]
```

```python
df
```

```python
df['item_price'] = df['item_price'].str.replace('$', '').astype(float)
df
```

```python
df[df["item_price"]>10]
```

### Step 5. What is the price of each item? 

print a data frame with only two columns `item_name` and `item_price`

```python
df["item_name"].unique()
```

```python
df['price_per_item'] = df['item_price'] / df['quantity']
unique_items = df[['item_name', 'price_per_item']].drop_duplicates().reset_index(drop=True)
unique_items.columns = ['item_name', 'item_price']
```

```python
unique_items
```

### Step 6. Sort by the name of the item

```python
unique_items_sorted = unique_items.sort_values(by='item_name').reset_index(drop=True)
unique_items_sorted
```
