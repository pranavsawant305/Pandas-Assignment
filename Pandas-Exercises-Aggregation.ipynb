### Step 4. Discover what is the mean age per occupation

```python
df = df['user_id|age|gender|occupation|zip_code'].str.split('|', expand=True)
df.columns = ['user_id', 'age', 'gender', 'occupation', 'zip_code']
df
```

```python
df["age"]=df["age"].astype(float)
```

```python
df.groupby(["occupation"])["age"].agg(["mean"])
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
df.groupby(["occupation"])["age"].agg(["min","max"])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
df.groupby(["occupation","gender"])["age"].agg(["mean"]).unstack()
```
